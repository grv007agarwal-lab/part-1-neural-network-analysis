##Task 4: Training and Evaluation

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import confusion_matrix, classification_report
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Input

# ---------------------------------
# 1. Load Dataset
# ---------------------------------
df = pd.read_csv("customer_churn_nn.csv")

# ---------------------------------
# 2. Remove Unnecessary Column
# ---------------------------------
df = df.drop("customer_id", axis=1)

# ---------------------------------
# 3. Encode Categorical Columns
# ---------------------------------
categorical_columns = [
    "region",
    "plan_type",
    "contract_type",
    "payment_method"
]

label_encoder = LabelEncoder()

for col in categorical_columns:
    df[col] = label_encoder.fit_transform(df[col])

# ---------------------------------
# 4. Separate Features and Target
# ---------------------------------
X = df.drop("churn", axis=1)
y = df["churn"]

# ---------------------------------
# 5. Feature Scaling
# ---------------------------------
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# ---------------------------------
# 6. Train-Test Split
# ---------------------------------
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled,
    y,
    test_size=0.2,
    random_state=42
)

# ---------------------------------
# 7. Build Neural Network
# ---------------------------------
model = Sequential()

# Input Layer
model.add(Input(shape=(X_train.shape[1],)))

# Hidden Layer 1
model.add(Dense(16, activation='relu'))

# Hidden Layer 2
model.add(Dense(8, activation='relu'))

# Output Layer
model.add(Dense(1, activation='sigmoid'))

# ---------------------------------
# 8. Compile Model
# ---------------------------------
model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

# ---------------------------------
# 9. Train Model
# ---------------------------------
history = model.fit(
    X_train,
    y_train,
    epochs=50,
    batch_size=32,
    validation_split=0.2,
    verbose=1
)

# ---------------------------------
# 10. Evaluate on Training Data
# ---------------------------------
train_loss, train_accuracy = model.evaluate(X_train, y_train)

print("Training Loss:", train_loss)
print("Training Accuracy:", train_accuracy)
print("\n")

# ---------------------------------
# 11. Evaluate on Testing Data
# ---------------------------------
test_loss, test_accuracy = model.evaluate(X_test, y_test)

print("Testing Loss:", test_loss)
print("Testing Accuracy:", test_accuracy)
print("\n")

# ---------------------------------
# 12. Predictions
# ---------------------------------
y_pred_prob = model.predict(X_test)

# Convert probabilities into 0 or 1
y_pred = (y_pred_prob > 0.5).astype(int)

# ---------------------------------
# 13. Confusion Matrix
# ---------------------------------
cm = confusion_matrix(y_test, y_pred)

print("Confusion Matrix:")
print(cm)
print("\n")

# ---------------------------------
# 14. Classification Report
# ---------------------------------
print("Classification Report:")
print(classification_report(y_test, y_pred, zero_division=0))
```

## Task 5: Hyperparameter Experimentation

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import accuracy_score
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Input

# ---------------------------------
# Load Dataset
# ---------------------------------
df = pd.read_csv("customer_churn_nn.csv")

# Drop ID column
df = df.drop("customer_id", axis=1)

# Encode categorical columns
categorical_columns = [
    "region",
    "plan_type",
    "contract_type",
    "payment_method"
]

label_encoder = LabelEncoder()

for col in categorical_columns:
    df[col] = label_encoder.fit_transform(df[col])

# Features and target
X = df.drop("churn", axis=1)
y = df["churn"]

# Scale features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled,
    y,
    test_size=0.2,
    random_state=42
)

# ---------------------------------
# Experiment Function
# ---------------------------------
def create_model(hidden_layers, neurons, activation):

    model = Sequential()

    # Input Layer
    model.add(Input(shape=(X_train.shape[1],)))

    # Hidden Layers
    for _ in range(hidden_layers):
        model.add(Dense(neurons, activation=activation))

    # Output Layer
    model.add(Dense(1, activation='sigmoid'))

    # Compile Model
    model.compile(
        optimizer='adam',
        loss='binary_crossentropy',
        metrics=['accuracy']
    )

    return model

# ---------------------------------
# Experiment Configurations
# ---------------------------------
experiments = [
    {
        "name": "Experiment 1",
        "hidden_layers": 1,
        "neurons": 16,
        "activation": "relu",
        "epochs": 30,
        "batch_size": 32
    },

    {
        "name": "Experiment 2",
        "hidden_layers": 2,
        "neurons": 32,
        "activation": "relu",
        "epochs": 50,
        "batch_size": 16
    },

    {
        "name": "Experiment 3",
        "hidden_layers": 3,
        "neurons": 64,
        "activation": "tanh",
        "epochs": 70,
        "batch_size": 16
    }
]

# ---------------------------------
# Run Experiments
# ---------------------------------
results = []

for exp in experiments:

    print(f"Running {exp['name']}")

    model = create_model(
        hidden_layers=exp["hidden_layers"],
        neurons=exp["neurons"],
        activation=exp["activation"]
    )

    # Train model
    model.fit(
        X_train,
        y_train,
        epochs=exp["epochs"],
        batch_size=exp["batch_size"],
        verbose=0
    )

    # Evaluate
    train_loss, train_accuracy = model.evaluate(
        X_train,
        y_train,
        verbose=0
    )

    test_loss, test_accuracy = model.evaluate(
        X_test,
        y_test,
        verbose=0
    )

    # Save results
    results.append({
        "Experiment": exp["name"],
        "Hidden Layers": exp["hidden_layers"],
        "Neurons": exp["neurons"],
        "Activation": exp["activation"],
        "Epochs": exp["epochs"],
        "Batch Size": exp["batch_size"],
        "Train Accuracy": round(train_accuracy, 4),
        "Test Accuracy": round(test_accuracy, 4),
        "Train Loss": round(train_loss, 4),
        "Test Loss": round(test_loss, 4)
    })

# ---------------------------------
# Display Results
# ---------------------------------
results_df = pd.DataFrame(results)

print("\nExperiment Comparison Table:\n")
print(results_df)
```

## Task 6: Final Reflection

**1. Role of Weights and Biases in a Neural Network**

Weights and biases are the core parameters that allow a neural network to learn patterns from data.

**Weights**

Weights determine the importance of each input feature.
During training, the neural network adjusts weights to minimize prediction error.

Neural computation:
$$z = \sum_{i=1}^{n} w_i x_i + b$$

Where:
* $x_i$ = input features
* $w_i$ = weights
* $b$ = bias
Bias shifts the output of the neuron.
It helps the model fit the data more flexibly.
Without bias, the model would have limited learning capability.

During backpropagation:

* Weights and biases are continuously updated to reduce loss and improve predictions.


---


**2. Why an Activation Function is Required**

Activation functions introduce non-linearity into the neural network.

Without activation functions:

* The neural network behaves like a simple linear model.
* Multiple layers would not provide additional learning power.

Example of ReLU activation:

f(x)=max(0,x)

Importance of Activation Functions
* Enable learning of complex patterns
* Help neural networks solve non-linear problems
* Improve model expressiveness

Common activation functions:

* ReLU
* Sigmoid
* Tanh


---


**3. Effect of Learning Rate**

The learning rate controls how much weights are updated during training.

Weight update equation:

$$W_{new} = W_{old} - \eta \frac{\partial L}{\partial W}$$

Where:
* $\eta$ = learning rate


**If Learning Rate is Too High**

* Model may overshoot the optimal solution.
* Training becomes unstable.
* Loss may fluctuate instead of decreasing.

**If Learning Rate is Too Low**

* Training becomes very slow.
* Model may take too long to converge.
* Can get stuck before reaching the best solution.

A balanced learning rate helps the model converge efficiently.


---

The model showed slight signs of overfitting in deeper experiments.

Evidence
* Training accuracy became higher than testing accuracy in larger models.
* Increasing layers and neurons improved training performance but not testing performance.

Example:

However, severe overfitting was not observed because:
